```python
import json
with open('bitcoin_price.json') as f:
    json_data = json.load(f)
```

```python
import sqlite3
connection = sqlite3.connect('bictoin_prices.db')
cursor = connection.cursor()
```

```python
json_data[1].items()
```

```python
query = """
DROP TABLE bitcoin_prices"""
cursor.execute(query)
```

```python
query = """
CREATE TABLE IF NOT EXISTS bitcoin_prices
(symbol text, time INTEGER, open real, close real,high real, low real, volume real)"""
cursor.execute(query)
```

```python
prices_data_list = []
for i in range(len(json_data)):
    list_prov = []
    for key,value in json_data[i].items():
        list_prov.append(value)
    tuple_prov = tuple(list_prov)
    prices_data_list.append(tuple_prov)
```

```python
prices_data_list
```

```python
query = """
INSERT INTO bitcoin_prices VALUES (?,?,?,?,?,?,?)
"""
cursor.executemany(query, prices_data_list)
```

```python
cursor.execute("SELECT * FROM bitcoin_prices LIMIT 2;")
cursor.fetchall()
```

```python
query = """
SELECT MIN(time)
FROM bitcoin_prices
"""
res = cursor.execute(query)
import datetime

min_date = res.fetchone()[0]

query = """
SELECT MAX(time)
FROM bitcoin_prices
"""
res = cursor.execute(query)

max_date = res.fetchone()[0]

print(datetime.datetime.fromtimestamp(min_date//1000))
print(datetime.datetime.fromtimestamp(max_date//1000))
```

```python
query = """
SELECT MAX(close),
STRFTIME('%Y', DATE(ROUND(time/1000), 'unixepoch')) as year
FROM bitcoin_prices
GROUP BY year
ORDER BY year
"""
cursor.execute(query)
cursor.fetchall()
```

```python
"""
Find the genre names with the longest average song length
"""
query = """
SELECT AVG(Milliseconds) as Duration,
genres.Name as Name
FROM tracks
JOIN genres ON tracks.GenreId = genres.GenreId
GROUP BY 2
ORDER BY 1 DESC
"""
cur.execute(query)
cur.fetchall()
```

```python
"""
TOTAL SALES FOR EACH YEAR AND GENRE"""

query = """
SELECT STRFTIME('%Y', i.InvoiceDate) as Year,
g.Name,
SUM(i.Total)

FROM invoices as i
JOIN invoice_items as i_i ON i.InvoiceId = i_i.InvoiceId
JOIN tracks as t on t.TrackId = i_i.TrackId
JOIN genres as g on g.GenreId = t.GenreId
GROUP BY 1,2
ORDER BY 1,2
"""
cur.execute(query)
res = cur.fetchall()
```

```python
import pandas as pd
data_frame = pd.DataFrame(res, columns=['year', 'genre','sales_USD'])
```

```python
data_frame.to_csv('chinook_data.csv', index= False)
data_frame.to_excel('chinook_data.xlsx', index= False)
```
